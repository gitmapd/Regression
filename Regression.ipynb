```python
import pandas as pd
import numpy as np
from sklearn.linear_model import LinearRegression
```

```python
data = {'Distancia':[5,6.5,7,7.7,8,8.6,9.4],'Calorias':[340,450,478,515,550,580,630]}
```

```python
df = pd.DataFrame(data)
```

```python
df.plot.scatter(x='Distancia', y='Calorias', title='Calorias por distancia')
```

```python
df.corr()
```

```python
df.describe()
```

```python
import matplotlib.pyplot as plt

dist = df['Distancia']
cals = df['Calorias']

plt.scatter(dist, cals)
plt.xlabel('Distancia')
plt.ylabel('Calorias')
```

```python
dist_np = dist.to_numpy()
cals_np = cals.to_numpy()

dist_np.shape, cals_np.shape
```

```python
from sklearn.linear_model import LinearRegression

sklearn_model = LinearRegression().fit(dist_np.reshape((-1,1)), cals_np)
sklearn_predict = sklearn_model.predict(dist_np.reshape((-1, 1)))
sklearn_predict.shape
```

```python
dist = df['Distancia']
cals = df['Calorias']

plt.scatter(dist, cals)
plt.xlabel('Distancia')
plt.ylabel('Calorias')

plt.scatter(dist, sklearn_predict)
```
